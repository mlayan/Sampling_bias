```python
##################################################
## Get all directories containing simulated data
##################################################
listFiles = []
listDirInput = []
listDirOutput = []

for root, subdirs, files in os.walk('.'): 
    print(root, subdirs, files)
    if 'files' in root:
        for f in files: 
            if 'fasta' in f:
                
                # Location of fasta and trait files
                dirInput = re.sub(r'^\./', '', root)
                
                # Location of XML files to generate
                dirOutput = re.sub('files', subDir, dirInput)
                testDirOuput = re.sub('files', subDir, root)
                if not Path(testDirOuput).exists():
                    os.mkdir(testDirOuput)
                
                # Lists of directories and file names
                listDirInput.append(dirInput)
                listDirOutput.append(dirOutput)
                listFiles.append(f)
```

```python
##################################################
## Get all directories containing simulated data
##################################################
listFiles = []
listDirInput = []
listDirOutput = []

# In a directory containing simulated data
# retrieve file names
for root, subDir, files in os.walk():     
    if 'files' in root: 
        n = int(re.match(r'simulation(\d+)', root).group(1))
        if n < 11:
            
            for f in files:
                if 'fasta' in f and 'stratified' not in f :
                    
                    # Location of fasta and trait files
                    dirInput = re.sub(r'^\./', '', root)

                    # Location of XML files to generate
                    dirOutput = re.sub('files', subDir, dirInput)
                    testDirOuput = re.sub('files', subDir, root)
                    if not Path(testDirOuput).exists():
                        os.mkdir(testDirOuput)

                    # Lists of directories and file names
                    listDirInput.append(dirInput)
                    listDirOutput.append(dirOutput)
                    listFiles.append(f)
```

```python
##################################################
# For each simulation produce the XML files 
# for all the sampling strategies and the  
# associated sh files to run BEAST
# Launch them in the meantime
##################################################
index = 0

for i in range(len(listFiles)):
    
    # File and directory
    files = listFiles[i]
    dirInput = listDirInput[i]
    dirOutput = listDirOutput[i]
    
    # Get trait file name
    regionsFile = re.sub('sequences', 'traits', files)
    regionsFile = re.sub('fasta', 'txt', regionsFile)
    
    # Tree file name and directory
    #treeFile= re.sub('sequences_', '', files)
    #treeFile = re.sub('fasta', 'nwk', treeFile)
    #treeFileDir = 'simulation' + re.match('sim(\d+)', files).group(1) + '/files'
    
    # Define the parameters logEvery and chainLength
    if re.search('150', files):
        logEvery = 2000
        chainLength = 20000000        
    else:
        logEvery = 4000
        chainLength = 40000000

    # Generate the xml file
    if subDir.lower() == "dta":
        xml_beast1(files, regionsFile, logEvery, chainLength, dirInput, dirOutput,
                   markovJumps = True
        #          outputName = "generator",
        #          fixedTreeFile=treeFile,
        #          fixedTreeDirectory=treeFileDir,
                   )
    
    if subDir.lower() == "mascot_v8":
        xml_mascot(files, regionsFile, logEvery, chainLength, 
                   version = 8,
                   dirInput = dirInput,
                   dirOutput = dirOutput
                  )
    #xml_basta(files, regionsFile, logEvery, chainLength, version = 3, outputName = 'test_v3')

print("Done")
```
